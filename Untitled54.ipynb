```python
# ‚úÖ Department-wise Monthly Attendance Tracker in Google Colab

import pandas as pd
import matplotlib.pyplot as plt

# -----------------------
# Step 1: Sample Dataset
# -----------------------
data = {
    "Date": ["2025-07-01", "2025-07-01", "2025-07-02", "2025-07-02", "2025-07-03", "2025-07-03"],
    "Department": ["CSE", "ECE", "CSE", "ECE", "CSE", "ECE"],
    "Present": [40, 35, 42, 37, 38, 30],
    "Total": [50, 50, 50, 50, 50, 50]
}

df = pd.DataFrame(data)

# Convert Date to datetime
df["Date"] = pd.to_datetime(df["Date"])

# -----------------------
# Step 2: Calculate Attendance %
# -----------------------
df["Attendance %"] = (df["Present"] / df["Total"]) * 100

print("üîπ Attendance Data")
print(df)

# -----------------------
# Step 3: Department-wise Average Attendance
# -----------------------
dept_avg = df.groupby("Department")["Attendance %"].mean().reset_index()

print("\nüîπ Department-wise Average Attendance")
print(dept_avg)

# -----------------------
# Step 4: Conditional Formatting (Low Attendance Warning)
# -----------------------
threshold = 75
low_attendance = df[df["Attendance %"] < threshold]

print("\n‚ö†Ô∏è Departments with Low Attendance (<75%)")
print(low_attendance)

# -----------------------
# Step 5: Visualization
# -----------------------

# 1Ô∏è‚É£ Line Chart - Date-wise trend per department
plt.figure(figsize=(8,5))
for dept in df["Department"].unique():
    subset = df[df["Department"] == dept]
    plt.plot(subset["Date"], subset["Attendance %"], marker="o", label=dept)

plt.title("Department-wise Attendance Trend")
plt.xlabel("Date")
plt.ylabel("Attendance %")
plt.legend()
plt.grid(True)
plt.show()

# 2Ô∏è‚É£ Pie Chart - Average attendance by department
plt.figure(figsize=(6,6))
plt.pie(dept_avg["Attendance %"], labels=dept_avg["Department"], autopct="%.1f%%")
plt.title("Department-wise Average Attendance")
plt.show()

# -----------------------
# Step 6: Save to Excel
# -----------------------
output_file = "Department_Attendance_Report.xlsx"
with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
    df.to_excel(writer, sheet_name="Daily Attendance", index=False)
    dept_avg.to_excel(writer, sheet_name="Dept Avg Attendance", index=False)
    low_attendance.to_excel(writer, sheet_name="Low Attendance Alerts", index=False)

print(f"\n‚úÖ Excel Report saved as {output_file}")
```
